```python
from fitter import Fitter 
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = stats.norm.rvs(loc=1.5, scale=2, size=1000)
```

```python
pd.Series(data).hist()
```

```python
f = Fitter(data,distributions=['norm'])
f.fit()
```

```python
f.summary(Nbest=10)
```

```python
f.fitted_param['norm']
```

```python
stats.kstest(data,cdf='norm', args=(1.5, 2))
```
